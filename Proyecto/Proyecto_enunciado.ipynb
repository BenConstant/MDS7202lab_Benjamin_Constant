```python
import os
import datetime
from   pathlib import Path
import re
from scipy.stats import zscore
class Profiler():
    columnas_sinNaN = []
    feature_anomala = []
    def __init__(self,dataframe):
        if isinstance(dataframe, pd.DataFrame):
            self.df        = dataframe
            self.fecha_hoy = datetime.datetime.now().strftime("%d_%m_%Y")
            self.nombre_carpeta_fecha ="EDA_" + self.fecha_hoy
            if not os.path.exists(self.nombre_carpeta_fecha):
                os.makedirs(self.nombre_carpeta_fecha)  
        else:
            raise TypeError(
                "Se debe entregar un archivo Dataframe")
            
    def summarize(self):
        df                   = self.df.copy()
        
        solo_numeros = df.select_dtypes(include=np.number)
        solo_str     = df.select_dtypes(include='object')
        
        with open(self.nombre_carpeta_fecha + '\summary.txt','w') as archivo:
    
            for nombre_column in solo_str.columns:
        
                df_new = pd.DataFrame(solo_str[nombre_column].fillna('Valores nulos').value_counts(dropna=False))
                anomalia = [indice for indice in df_new.index if len(re.split(r'(?![\d+.\d+])[^A-Za-z0-9]',indice))>2]
                if anomalia:
                    print('El feature '+nombre_column+' presenta simbolos poco comunes en los datos')
                    self.feature_anomala.append(nombre_column)
                    
                df_new.loc['total_unicos'] = df_new[df_new==1].sum()
                df_new.loc['dtype column'] = solo_str[nombre_column].dtype
                
                if not 'Valores nulos' in df_new.index:
                    df_new.loc['Valores nulos'] = 0
                    self.columnas_sinNaN.append(nombre_column)
                    
                archivo.write('----------Columnna Dataset: ' + nombre_column + '-----------------------\n')
                archivo.write(' \n')
                df_new.loc[['Valores nulos','total_unicos','dtype column'],:].to_csv(archivo,sep='\t',index=True)
                archivo.write(' \n')
            
            for nombre_column in solo_numeros.columns:
                
                df_new = pd.DataFrame(solo_numeros[nombre_column].describe())
                df_new.loc['Negativo'] = solo_numeros[nombre_column][solo_numeros[nombre_column]<0].sum()
                df_new.loc['Cero']     = solo_numeros[nombre_column][solo_numeros[nombre_column]==0].sum()
                q1 = solo_numeros[nombre_column].quantile(0.25)
                q3 = solo_numeros[nombre_column].quantile(0.25)
                iqr = q3-q1
                umbral_inf = q1 - 1.5*iqr
                umbral_sup = q3 + 1.5*iqr
                try:
                    df_new.loc['outliers'] = len(solo_numeros[nombre_column][(solo_numeros[nombre_column] < umbral_inf)|(solo_numeros[nombre_column]>umbral_sup)].dropna())
                except ValueError:
                    df_new.loc['outliers'] = 0
                    
                archivo.write('----------Columnna Dataset: ' + nombre_column + '-----------------------\n')
                archivo.write(' \n')
                df_new.to_csv(archivo,sep='\t',index=True)
                archivo.write(' \n')
      
    def plot_vars(self,variables_max=7):
        self.path_plots = self.nombre_carpeta_fecha + '\plots'
        
        if not os.path.exists(self.path_plots):
                os.makedirs(self.path_plots)
        df                   = self.df.copy()
        solo_numeros = df.select_dtypes(include=np.number)
        solo_str     = df.select_dtypes(include='object')   
        
        numeros = pd.DataFrame(solo_numeros.value_counts())
        fig, axs= plt.subplots(figsize=(7,4),tight_layout=True,facecolor='w')
        axs.hist(solo_numeros)
        axs.set_ylabel('Count [-]',fontsize=15,fontname='serif')
        axs.set_xlabel('A√±os', fontsize=15,fontname='serif')
        axs.set_title('Years',fontsize=15,fontname='serif')
        plt.savefig(self.path_plots+ "\correlacion_numero.pdf", format="pdf",bbox_inches='tight')
        
        for nombre_column in solo_str.columns:
            
            df_new           = solo_str[nombre_column].value_counts().sort_values()
            filtro_categoria = df_new.head(variables_max).index 
            df_new_filtrado = solo_str[solo_str[nombre_column].isin(filtro_categoria)]

            df_plot = pd.DataFrame(df_new_filtrado[nombre_column].value_counts())
            
            fig, ax = plt.subplots(figsize=(7,4),tight_layout=True,facecolor='w')
            ax.bar(df_plot.index,df_plot[nombre_column])
            ax.set_ylabel('Count [-]',fontsize=10,fontname='serif')
            name_resumen = ['\n'.join(texto.split(' ')) for texto in df_plot.index]
            ax.set_xticklabels(name_resumen,rotation=45,fontsize=7)
            for tick in ax.xaxis.get_major_ticks():
                tick.set_pad(20)
    
            plt.savefig(self.path_plots+ str("\\") + nombre_column+ ".pdf", format="pdf",bbox_inches='tight')
        enc = OrdinalEncoder()
        ordinales = enc.fit_transform(solo_str.dropna())
        
        correlacion = pd.DataFrame(ordinales,columns=solo_str.columns).corr()
        
        fig, ax = plt.subplots(figsize=(9, 7))

        heatmap = ax.imshow(correlacion, cmap='viridis', vmin=-1, vmax=1, aspect='auto')
        cbar = fig.colorbar(heatmap)
        ax.set_xticklabels(correlacion.columns)
        ax.set_yticklabels(correlacion.columns)
        plt.setp(ax.get_xticklabels(), rotation=45, ha="right", rotation_mode="anchor")
        plt.title("Correlaci√≥n entre Variables")
        plt.savefig(self.path_plots+ "\correlacion_cat.pdf", format="pdf",bbox_inches='tight')
        
    def clean_data(self,all_drop=False):
        self.path_clean = self.nombre_carpeta_fecha + '\clean_data'
        
        if not os.path.exists(self.path_clean):
                os.makedirs(self.path_clean)
                
        df           = self.df.copy()
        solo_numeros = df.select_dtypes(include=np.number)
        solo_str     = df.select_dtypes(include='object')
        
        drop_df_total  = df.dropna()
        
        enc             = OrdinalEncoder()
        ordinales       = enc.fit_transform(solo_str.dropna())
        correlacion_str = pd.DataFrame(ordinales,columns=solo_str.columns).corr()
        #corre_filtro    = pd.DataFrame(correlacion_str.sum())[correlacion_str.sum()<0.5].drop(self.columnas_sinNaN)
        #if not len(corre_filtro)<1:
        #    solo_str_filtro = solo_str[solo_str[corre_filtro.index].notna()]
   
        pipeline_final = Pipeline([
                ('limpiar_separacion',FunctionTransformer(func=limpiar_separacion)),
                ('Separar',FunctionTransformer(func=separar))
                ])
    
        Transf_conv    = FunctionTransformer(func=cambio_float)
        
        df[['age','height','weight']] = pipeline_final.fit_transform(df['age-height-weight'])

        df['age']    = Transf_conv.transform(df['age'])
        df['height'] = Transf_conv.transform(df['height'])
        df['weight'] = Transf_conv.transform(df['weight'])
        
        df = df.drop(columns='age-height-weight')
        
        df = df[df['age'].notna()]
        self.path_data = self.path_clean + '\data.csv'
        if all_drop:
            drop_df_total.to_csv(self.path_clean + '\data.csv')
        else:
            df.to_csv(self.path_clean + '\data.csv')
        
            
    def scale(self,scaler=MinMaxScaler(),numTransf=PowerTransformer(method='yeo-johnson'),Imputer=SimpleImputer(strategy="constant", fill_value="missing"),catTransf=OrdinalEncoder()):
        self.path_scale = self.nombre_carpeta_fecha + '\scale'
        if not os.path.exists(self.path_scale):
                os.makedirs(self.path_scale)
                
        df = pd.read_csv(self.path_data).set_index('ID')
        solo_numeros = df.select_dtypes(include=np.number)
        solo_str     = df.select_dtypes(include='object')
        
        number = Pipeline([('scaler', scaler),
                    ('yeo-johnson', numTransf) ])
        categ = Pipeline(
                steps=[
                        ("imputer_cat", Imputer),
                        ("OrdinalEncoder", catTransf),
                        ])
        
        preprocessor = ColumnTransformer(
                        transformers=[
                            ('number',number,solo_numeros.columns),
                            ('Categ',categ,solo_str.columns)
                            ])
        df_new = pd.DataFrame(preprocessor.fit_transform(df),columns=df.columns)

        self.path_scale_df = self.path_scale + '\scaled_features.csv'
        df_new.to_csv(self.path_scale_df)
        
    def make_clusters(self,clustering = KMeans):
        self.path_cluster = self.nombre_carpeta_fecha + '\clusters'
        if not os.path.exists(self.path_cluster):
                os.makedirs(self.path_cluster)
        df           = pd.read_csv(self.path_scale_df)
        
        intertias = [
                [i,KMeans(n_clusters=i,random_state=0).fit(df.dropna()).inertia_]
                for i in range(2,20)
                ]
        fig, ax = plt.subplots(figsize=(9, 7))
        intertias = pd.DataFrame(intertias, columns=["n¬∞ clusters", "inertia"])
        ax.plot(intertias['n¬∞ clusters'],intertias['inertia'])
        ax.set_xlabel('n cluster')
        ax.set_ylabel('intertia')
        
        plt.savefig(self.path_cluster + "\codos.pdf", format="pdf",bbox_inches='tight')
        #px.line(intertias,x="n¬∞ clusters",y="inertia",title="M√©todo del Codo",height=600,            )
        
        # Por el largo de los datos y mi notebook, reducire el df para que corra el modelo mas rapido
        tsn  = TSNE(n_components=2, random_state=0)
        df_drop = df.dropna()
        df_short = df_drop[:20000]
        df_tsne = tsn.fit_transform(df_short)
        kmean = clustering(n_clusters=8,random_state=0).fit(df_short)
        df_short['Cluster'] = kmean.labels_
        df_short['x_tsne']  = df_tsne[:,0]
        df_short['y_tsne']  = df_tsne[:,1]
        fig, ax = plt.subplots(figsize=(9, 7))
        clust = ax.scatter(df_short['x_tsne'],df_short['y_tsne'],c=df_short['Cluster'])
        cbar = fig.colorbar(clust)
        ax.set_xlabel('x_tsne')
        ax.set_ylabel('y_tsne')
        plt.savefig(self.path_cluster + "\cluster.pdf", format="pdf",bbox_inches='tight')
        df_short.to_csv(self.path_cluster +'\data_clusters.csv')
        
    def detect_anomalies(self,metodo=zscore()):
        self.path_anomalias = self.nombre_carpeta_fecha + '\Anomalies'
        if not os.path.exists(self.path_anomalias):
                os.makedirs(self.path_anomalias)
        df   = pd.read_csv(self.path_scale_df)
        df   = df.dropna()
        anomalia_pipe = Pipeline([
            ('z_score',metodo)
        ],remainder='passthrough')
        preproccess = ColumnTransformer([
            ('all_values',anomalia_pipe,df.columns)
        ])
        df_anom = pd.DataFrame(preproccess.fit_transform(df),columns=df.columns)
        df_anom.to_csv(self.path_anomalias + '\data_anomalies.csv')
        
    def profile(self):
        self.summarize()
        self.plot_vars()
        self.clean_data()
        self.scale()
        self.make_clusters()
        self.detect_anomalies()
        
    def clearGarbage(self):
        for filename in os.listdir():
            if not len(filename.split('.'))>1:
                # Este codigo funciona si se tiene permiso para controlar la carpeta, sino sale un error WinErrir 5
                os.remove(os.path.join(Path.cwd(),filename))
    
def limpiar_separacion(Serie):
    return Serie.apply(lambda x: '-'.join(re.split(r'(?![\d+.\d+])[^A-Za-z0-9]',x)))
def cambio_float(Serie):
    return Serie.apply(lambda x: float(x) if x != 'nan' else np.nan)
def separar(Serie):
    return Serie.str.split('-',expand=True)
def log_transform(x):
    return np.log(x + 1)       
```
