```python
def limpiar_separacion(Serie):
    return Serie.apply(lambda x: '-'.join(re.split(r'(?![\d+.\d+])[^A-Za-z0-9]',x)))
def separar(Serie):
        return Serie.str.split('-',expand=True)
def cambio_float(Serie):
    return Serie.apply(lambda x: float(x) if x != 'nan' else np.nan)

pipeline_final = Pipeline([
    ('limpiar_separacion',FunctionTransformer(func=limpiar_separacion)),
    ('Separar',FunctionTransformer(func=separar))
])
dataframe[['age','height','weight']] = pipeline_final.fit_transform(dataframe['age-height-weight'])
dataframe['age'] =dataframe['age'].apply(lambda x: float(x) if x != 'nan' else np.nan)
dataframe['height'] =dataframe['height'].apply(lambda x: float(x) if x != 'nan' else np.nan)
dataframe['weight'] =dataframe['weight'].apply(lambda x: float(x) if x != 'nan' else np.nan)
```

```python
df = pd.read_csv('EDA_10_11_2023\clean_data\data.csv').set_index('ID')
solo_numeros = df.select_dtypes(include=np.number)
solo_str     = df.select_dtypes(include='object')

#preprocessing = ColumnTransformer(transformers=[
#    ()
#])
```

```python
number = Pipeline([('scaler', MinMaxScaler()),
                    ('yeo-johnson', PowerTransformer(method='yeo-johnson')) ])
number.fit_transform(solo_numeros)
```

```python
categ = Pipeline(
    steps=[
        ("imputer_cat", SimpleImputer(strategy="constant", fill_value="missing")),
        ("OrdinalEncoder", OrdinalEncoder()),
    ]
)
preprocessor = ColumnTransformer(
                transformers=[
                    ('number',number,solo_numeros.columns),
                    ('Categ',categ,solo_str.columns)
                ]
)
df_new = pd.DataFrame(preprocessor.fit_transform(df),columns=df.columns)
```

```python
df_new
```

```python
kmean = KMeans(n_clusters=8,random_state=0).fit(df_new.dropna())
```

```python
#from MulticoreTSNE import MulticoreTSNE
from sklearn.decomposition import PCA
#tsn= TSNE(n_components=2,random_state=0)
tsn  = TSNE(n_components=2, random_state=0)
pca  = PCA(n_components=7,svd_solver='full')
df_new = df_new.dropna()
df_pca = pd.DataFrame(pca.fit_transform(df_new))
df_pca
```

```python
kmean = KMeans(n_clusters=8,random_state=0).fit(df_pca[:20000])
```

```python
df_tsne = tsn.fit_transform(df_pca[:20000])
```

```python
df_final = df_pca[:20000]
```

```python
import plotly.graph_objects as go
import plotly

df_final['Cluster'] = kmean.labels_
df_final['x_tsne']  = df_tsne[:,0]
df_final['y_tsne']  = df_tsne[:,1]
fig = px.scatter(df_final,x='x_tsne',y='y_tsne',color='Cluster')
fig.show()
```

```python
fig.write_image('Example.png')
```
